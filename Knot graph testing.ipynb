```python
import snappy
def view(link):
    snap = {c: snappy.Crossing(c.label) for c in link.crossings}
    for c in link.crossings:
        for i, cs in enumerate(c.adjacent):
            snap[c][i] = cs
    _L = snappy.Link(list(snap.values()))
    _L.view()      # this doesn't open the plink window for some reason

    #L_pd = snappy.Link(link.PD_code())
    #L_pd.view()
```
