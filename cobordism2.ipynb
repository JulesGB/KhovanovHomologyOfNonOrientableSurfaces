```python
import snappy
from khovanov import *
from copy import deepcopy
from copy import copy
```

```python
def view(link):
    _snap = {c: snappy.Crossing(c.label) for c in link.crossings}
    for c in link.crossings:
        for i, cs in enumerate(c.adjacent):
            # _cs = deepcopy(cs)
            _cs = copy(cs)
            _snap[c][i] = _cs
    _L = snappy.Link(list(_snap.values()))
    _L.view()      # this doesn't open the plink window for some reason
```

```python
def view_movie(cob):
    i = 0
    for link in cob.links:
        print(i)
        i += 1
        view(link)
```
