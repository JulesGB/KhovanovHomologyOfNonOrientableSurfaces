# Cobordism Induced Map Computations

```python
def draw_movie(M):
    for l in M.links:
        draw_adjacency_graph(l)
        plt.show()
```

## $T_{2,5}$ to Unknot $\mathcal{U}$

```python
T2_5 = Link([(1,9,2,8),(7,3,8,2),(10,4,11,3),(4,12,5,11),(12,6,1,5),(6,10,7,9)])
```

```python
S = Cobordism(T2_5)
S.band_move(0,(4,1),(0,1))
S.finish()
```

```python
compare(S.KJ_class(), CKhElement([]))
```

The resulting KJ class is trivial, meaning the band move reduces $T_{2,5}$ to the unknot $\mathcal{U}$.

## Slice disk movie graph printing 

```python
def draw_slice_movies(S, T):
    n = len(S.links)
    for i,s,t in zip(range(n), S.links, T.links):
        plt.figure(i, figsize=[10,5*n])
        s_move = 'inital knot' if i==0 else S.movie[i-1][0] + ' applied'
        t_move = 'inital knot' if i==0 else T.movie[i-1][0] + ' applied'
        plt.subplot(121, title=f'S movie ({s_move})')
        draw_adjacency_graph(s, rad_increment=0.1)
        plt.subplot(122, title=f'T movie ({t_move})')
        draw_adjacency_graph(t, rad_increment=0.1)
        plt.show()
```
