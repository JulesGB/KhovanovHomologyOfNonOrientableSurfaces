## Induced Map Computation

```python
# T(2,5)
PD_code = [(5,1,6,10),(1,7,2,6),(7,3,8,2),(3,9,4,8),(9,5,10,4)]
K = Link(PD_code)
```

```python
snappy.Link(PD_code).view(show_crossing_labels=True)
```

```python
S = Cobordism(K)
S.band_move(-1, (0,2), (2,1))
S.finish()
S
```

```python
for link in S.links:
    snappy.Link(link.PD_code()).view(show_crossing_labels=True)
```
