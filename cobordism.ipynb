## Induced Map Computation

```python
# T(2,5)
PD_code = [(5,1,6,10),(1,7,2,6),(7,3,8,2),(3,9,4,8),(9,5,10,4)]
K = Link(PD_code)
```

```python
print(K)
```

```python
snappy.Link(PD_code).view(show_crossing_labels=True)
```

```python
S = Cobordism(K)
print(S0)
print('--------')
S.band_move(-1, (0,1), (2,1))
print(S0)
print('--------')
#S.finish()
#print(S)
```

```python
for link in S.links:
    snappy.Link(link.PD_code()).view(show_crossing_labels=True)
```

```python
S.KJ_class()
```

```python
L = Link([(9, 4, 10, 5), (5, 8, 6, 9), (11, 2, 12, 3), (3, 10, 4, 11), (1, 7, 2, 6), (7, 1, 8, 12)])
S0 = Cobordism(L)
print(S0)
print('--------')
S0.band_move(-1, (0, 0), (2, 1))
print(S0)
print('--------')
S0.finish()
print(S0)
print('--------')

S1 = Cobordism(L)
S1.band_move(-1, (1, 2), (3, 3))
S1.finish()
```

```python
compare(S0.KJ_class(), S1.KJ_class())
```

```python
compare(S0.mirror().KJ_class(), S1.mirror().KJ_class())
```
