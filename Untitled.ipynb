```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
data = np.genfromtxt("/home/jules/Downloads/habitat raw binned - habitat.csv", delimiter=",", skip_header=1)
```

```python
agri = data[:,0]
settled = data[:,1]
rural = data[:,2]
```

```python
# Remove duplicates
agri = agri[~np.isnan(agri)]
settled = settled[~np.isnan(settled)]
rural = rural[~np.isnan(rural)]
```

```python
res = scipy.stats.tukey_hsd(agri,settled,rural)
```

```python
labels = ["agri", "settled", "rural"]
```

```python
k = len(labels)

# Extract matrices
stat = np.array(res.statistic)   # shape (3,3)
pval = np.array(res.pvalue)
lower_mat, upper_mat = res.confidence_interval()   # each shape (3,3)

pairs = []
diffs = []
lower_err = []
upper_err = []

# Extract upper triangle pairs
for i in range(k):
    for j in range(i+1, k):
        pairs.append(f"{labels[i]} vs {labels[j]}")
        
        d = stat[i, j]
        diffs.append(d)

        lo = lower_mat[i, j]
        hi = upper_mat[i, j]

        lower_err.append(d - lo)
        upper_err.append(hi - d)

diffs = np.array(diffs)
lower_err = np.array(lower_err)
upper_err = np.array(upper_err)

# ---- Plot ----
x = np.arange(len(diffs))
plt.figure(figsize=(7,4))

plt.errorbar(
    x,
    diffs,
    xerr=[lower_err, upper_err],
    fmt="o",
    markersize=6,
    capsize=6
)

plt.axhline(0, color="gray", linestyle="--")

plt.yticks(x, pairs, rotation=20)
plt.xlabel("Mean difference")
plt.title("Tukey HSD Pairwise Comparisons")

plt.tight_layout()
plt.show()
```
